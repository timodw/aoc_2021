# Imports

```python
import numpy as np
from time import time
```

# Load input

```python
#INPUT_FILE = 'test.txt'
INPUT_FILE = 'input.txt'
```

```python
input_list = [[c for c in line.strip()] for line in open(INPUT_FILE, 'r')]
input_list
```

# Part 1

```python
start_t = time()
points_map = {
    ')': 3,
    ']': 57,
    '}': 1197,
    '>': 25137
}

opening_map = {
    ')': '(',
    ']': '[',
    '}': '{',
    '>': '<'
}

opening_set = set(('(', '[', '{', '<'))

stack = []
solution = 0
for line in input_list:
    for c in line:
        if c in opening_set:
            stack.append(c)
        else:
            if stack[-1] == opening_map[c]:
                stack.pop()
            else:
                solution += points_map[c]
                break
end_t = time()
print(f'SOLUTION: {solution}, TIME: {round((end_t - start_t) * 1000, 4)}ms')
```

# Part 2

```python
start_t = time()
points_map = {
    ')': 1,
    ']': 2,
    '}': 3,
    '>': 4
}

closing_map = {
    '(': ')',
    '[': ']',
    '{': '}',
    '<': '>'
}

opening_set = set(('(', '[', '{', '<'))

point_list = []
for line in input_list:
    valid_line = True
    stack = []
    for c in line:
        if c in opening_set:
            stack.append(c)
        else:
            if stack[-1] == opening_map[c]:
                stack.pop()
            else:
                valid_line = False
                break
    if valid_line:
        line_points = 0
        for c in stack[::-1]:
            line_points *= 5
            line_points += points_map[closing_map[c]]
        point_list.append(line_points)
solution = int(np.median(point_list))
end_t = time()
print(f'SOLUTION: {solution}, TIME: {round((end_t - start_t) * 1000, 4)}ms')
```
